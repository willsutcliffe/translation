# Performance of various trainings

## Russian English Training

### English to Russian

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

df1=  pd.read_csv('New_en-ru-training_info_after_6epochs_2m.csv')
df2=  pd.read_csv('New_en-ru-training_info_after_12epochs_2m.csv')
df3=  pd.read_csv('New_en-ru-training_info_after_16epochs_2m.csv')
df4=  pd.read_csv('New_en-ru-training_info_after_24epochs_2m.csv')
```

```python
df = pd.concat([df1,df2,df3,df4])
```

```python
df.head()
```

```python
epochs=np.arange(1,len(df)+1)
plt.plot(epochs,df.Train_loss,label='Train')
plt.plot(epochs,df.Val_loss,label='Val')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
```

```python
plt.plot(epochs,df.Val_BERT_score)
plt.xlabel('Epochs')
plt.ylabel('BERT')
```

### Russian to English Training

```python
df1=  pd.read_csv('New_ru-en-training_info_after_6epochs_2m.csv')
df2=  pd.read_csv('New_ru-en-training_info_after_13epochs_2m.csv')
df = pd.concat([df1,df2])
```

```python
epochs=np.arange(1,len(df)+1)
plt.plot(epochs,df.Train_loss,label='Train')
plt.plot(epochs,df.Val_loss,label='Val')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.text(15,2.5,'Russian -> English')
plt.legend()
plt.savefig('Loss_ru_en.png',dpi=300)
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Val BLEU', color=color)
ax1.plot(epochs, df.Val_BLEU_score, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Val BERT', color=color)  # we already handled the x-label with ax1
ax2.plot(epochs, df.Val_BERT_score, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.text(12,0.775,'Russian -> English')
plt.legend()
plt.savefig('Perf_ru_en.png',dpi=300)
```
